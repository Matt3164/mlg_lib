# imgproc
> Image processing utilities

```python
# export
from enum import Enum 

class ImageFormat(Enum):
    OPENCV = "opencv"
    SKIMAGE = "skimage"
    NUMPY="numpy"
```

```python
# export

from pathlib import Path
import numpy as np
import cv2
from skimage.io import imread

def load_img(path: Path, img_format: ImageFormat):
    if img_format==ImageFormat.OPENCV:
        return cv2.imread(str(path))
    elif img_format==ImageFormat.SKIMAGE:
        return imread(str(path))
    elif img_format==ImageFormat.NUMPY:
        return np.load(str(path))
    else:
        raise NotImplementedError
        
```

```python
# export

from pathlib import Path
import numpy as np
import cv2
from skimage.io import imsave

def save_img(arr: np.ndarray, path: Path, img_format: ImageFormat):
    if img_format==ImageFormat.OPENCV:
        return cv2.imwrite(str(path), arr)
    elif img_format==ImageFormat.SKIMAGE:
        return imsave(str(path), arr)
    elif img_format==ImageFormat.NUMPY:
        return np.save(str(path), arr)
    else:
        raise NotImplementedError
        
```

```python
for img_format, ext in zip([ImageFormat.NUMPY, ImageFormat.OPENCV, ImageFormat.SKIMAGE], ["npy", "jpg", "jpg"]):
    
    im_path = "/tmp/chelsea.{}".format(ext)
    
    im = chelsea()

    save_img(im, Path(im_path), img_format)

    loaded_im = load_img(Path(im_path), img_format=img_format)

    imshow(loaded_im)
    axis("off")
```
